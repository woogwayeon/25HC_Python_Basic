    💚 try-except 문 : 오류 처리를 위한 구문
    try 블록 수행중 오류가 발생하면 except 블록 수행
    try 블록에서 오류가 발생하지 않으면 except 블록 미수행

```python
try:
    4/0
except ZeroDivisionError as e:
    print(e)
    print("0으로 나눌 수 없습니다")
```

```python
try:
    a = [1,2]
    print(a[3])
    5/0

except ZeroDivisionError as e:
    print(f"{e} : 0으로 나눌 수 없습니다")

except IndexError as e:
    print(f"{e} : 해당하는 인덱스가 없습니다")
```

```python
# try-else 문
try:
    inp = int(input("나이를 입력하세요\n"))
    age = inp

except:
    # 오류 명시 안하면 바로 except 구문으로 빠짐
    print("숫자만 입력하세요")

else:
    if age < 18 :
        print("미성년자는 출입할 수 없습니다")
    else :
        print("왕좌의게임 시청가능!")
```

```python
# 오류 회피
try:
    f = open("없는 파일입니다", 'r')

except FileNotFoundError :
    pass
    # pass를 해두면 오류 그냥 회피함
```

```python
# 여러가지 예외처리
def risky():
    return 5/0

def risky2():
    a = [1,2]
    return a[2]

try:
    #risky()
    risky2()
except Exception as e:
    print(f"{e} : {type(e)} 에러가 발생했습니다")
```

```python
# re-raise

def risky3():
    raise ValueError("잘못된 입력입니다!")

def wrapper():
    try:
        risky3()

    except ValueError as e:
        print(f"에러 발생 : {e}")
        raise RuntimeError("wrapper 단계에서 실패했습니다") from e

wrapper()
```

```python
# 예외 만들기

class MyError(Exception):
    pass

def sayNick():
    nick = input()
    if nick == '바보':
        try:
            raise MyError()
        except Exception as e:
            print(f"바보는 안돼")
    else:
        print(f"닉네임은 {nick} 입니다")


sayNick()
```

```python
class MyError(Exception):
    def __str__(self):
        return "허용되지 않은 별명입니다"

def sayNick(nick):
    if nick == '바보':
        raise MyError()
    print(nick)

try:
    sayNick('바보')
    sayNick('천사')
except MyError as e:
    print(e)
```
