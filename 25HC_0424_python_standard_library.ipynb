```python
import time
time.time() #1970년 1월 1일 이후로 흘러간 시간
```

```python
time.localtime(time.time()) # 현재 시간정보 출력
```

```python
print(time.asctime())  # 아스키타임
print(time.ctime())    # current 현재시간
```

```python
t1 = time.strftime('%x', time.localtime())
t2 = time.strftime('%c', time.localtime())

print(t1)
print(t2)
```

```python
# 오늘운 몇월 멸일 입니다 해보자
now = time.localtime()

n2 = time.strftime("%Y년 %m월 %d일", now)
# %Y : 4자리 연도, %m : 2자리 월, %d : 2자리 일
print(f"오늘은 {n2}입니다\n")

filename = "log_"+time.strftime("%Y%m%d", now)+".txt"
print(f"{filename}\n")

alram = time.strftime("%A", now)
print(f"알람은 {alram}에 울립니다\n")
```

```python
with open(filename, 'w') as f:
    f.write("귀도 반 로섬이 만들었습니다 파이선을")
```

## 최대공약수 (GCD)
Greatest Common Divisor

어린이집에서 사탕 60개, 초콜릿 100개, 젤리 80개를 준비했다!

아이들이 서로 싸우지 않도록 나누어 봉지에 담는다고 하면 몇 봉지까지 가능할까?

```python
import math

div_candy = math.gcd(60, 100, 80)
print(div_candy)

working_date = math.lcm(3,9,6)
print(working_date)
```

어느 버스 정류장에 시내버스는 15분마다 도착하고 마을버스는 25분마다 도착한다고 한다!

오후 1시에 두 버스가 동시에 도착했다고 할때 두 버스가 동시에 도착할 다음시각을 알려
면 어떻게 해야할까?

```python
bus = math.lcm(15, 25)
print(bus)
```

## 가위바위보 게임 만들기

```python
import random
import sys

rsp = ["가위", "바위", "보"]
cpt = random.choice(rsp)

print("\n가위 바위 보 중에 하나를 입력하세요\n")
user = input()

if not user in rsp :
    print("잘못된 입력입니다")
    sys.exit()

print(f"\n컴퓨터 : {cpt} / 사용자 : {user}\n")

if user==cpt:
    print("비겼습니다")
elif (user == rsp[0] and cpt == rsp[2]) or \
     (user == rsp[1] and cpt == rsp[0]) or \
     (user == rsp[2] and cpt == rsp[1]):
    print("사용자가 이겼습니다!")
else:
    print("컴퓨터가 이겼습니다")


```

### itertools.zip_longest

image.png

```python
import itertools

stu = ['서연', '아인', '정근', '교수님', '철수']
snaks = ['사탕', '초콜릿', '젤리']

res = itertools.zip_longest(stu, snaks, fillvalue="오레오")
                            # 설문조사같은거 받을 때 값이 없는 부분에 뭐 채워서 데이터 정제할지 사용된다

# iterable한 자료 자체이기 때문에 자료형을 정해서 출력해야 한다!
# print(list(res))
# print(tuple(res))
print(set(res)) # set을 사용할때는 값 중복이 안되기 때문에 조심해서 씁시다!
```
